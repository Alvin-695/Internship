```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

##### 1) Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
page= requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
soup= BeautifulSoup(page.content)
```

```python
header=[]

for i in soup.find_all("span",class_="mw-headline"):
    header.append(i.text)

df = pd.DataFrame(header, columns=["Header"])
print(df)
```

##### 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame i) Headline ii) Time iii) News Link

```python
web=requests.get("https://www.cnbc.com/world/?region=world")
web
```

```python
page=BeautifulSoup(web.content)
```

```python
news=page.find_all("div",class_="LatestNews-container")
```

```python
headline=[]
time=[]
news_link=[]

for i in news[0:16]:
    headline.append(i.find("a").text.strip())
    time.append(i.find("time",class_="LatestNews-timestamp").text.strip())
    news_link.append(i.find('a')['href'])
    
d={"Headline":headline,"Time":time,"News_Link":news_link}
df=pd.DataFrame(d)
df
```

##### 6) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-i) Paper Title ii) Authors iii) Published Date iv) Paper URL

```python
page=requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
article=soup.find("ul",class_="sc-9zxyh7-0 cMKaMj")
```

```python
paper_title=[]
authors=[]
published_date=[]
paper_url=[]

for i in article:
    paper_title.append(i.find("h2",class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg").text.strip())
    authors.append(i.find("span",class_="sc-1w3fpd7-0 dnCnAO").text.strip())
    published_date.append(i.find("span",class_="sc-1thf9ly-2 dvggWt").text.strip())
    paper_url.append(i.find("a")['href'])
    
d={"Paper_title":paper_title,"Authors":authors,"Published_Date":published_date,"Paper_URL":paper_url}
df=pd.DataFrame(d)
df
```

##### 7) Write a python program to scrape mentioned details from dineout.co.in and make data frame-i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL

```python
page=requests.get("https://www.dineout.co.in/jaipur-restaurants/welcome-back")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
swiggy=soup.find("div",class_="restnt-card-wrap-new")
```

```python
restaurant_name=[]
cuisine=[]
location=[]
ratings=[]
image_url=[]

for i in swiggy:
    restaurant_name.append(i.find("a",class_="restnt-name ellipsis").text.strip())
    cuisine.append(i.find("span",class_="double-line-ellipsis").text.strip())
    location.append(i.find("div",class_="restnt-loc ellipsis").text.strip())
    ratings.append(i.find("div",class_="restnt-rating").text.strip())
    image_url.append(i.find("img",class_="no-img")['data-src'])
    
d={"Restaurant_name":restaurant_name,"Cuisine":cuisine,"Location":location,"Ratings":ratings,"Image_URL":image_url}
df=pd.DataFrame(d)
df
```
