##### 1&2  Write a python program which searches all the product under a particular product from www.amazon.in. The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars. 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "BrandName", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
amazon=webdriver.Chrome()
amazon.get('http://www.amazon.in') #Starting chrome webdriver
```

```python
#Locating the search bar input the product
product=amazon.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
product.send_keys(input('Enter product name:'))
search_button=amazon.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
search_button.click() 
```

```python
 # Finding all the products in the search results
urls=[]
for page in range(0,3):
    url=amazon.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        urls.append(i.get_attribute('href'))# Getting the product url
next_button=amazon.find_elements(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[29]/div/div/span/a[3]')
next_button.click()
len(urls)
```

```python
# Extracting the relevant information for each product
brand=[]
product_name=[]
expected_delivery=[]
availability=[]
price=[]
return_exchange=[]


for i in urls:
    amazon.get(i)
    time.sleep(1)
    try:
        br=amazon.find_elements(By.XPATH,'//td[@class="a-span9"]') # Getting the product brand
        brand.append(br.text)
    except NoSuchElementException :
        brand.append('-')

    try:
        product=amazon.find_elements(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        product_name.append(product.text) #Getting the product title
    except NoSuchElementException :
        product_name.append('-')

    try:
        pr=amazon.find_elements(By.XPATH,'//span[@class="a-price-whole"]')# Getting the product price
        price.append(pr.text)
    except NoSuchElementException :
        price.append('-')
        
    try:
        re=amazon.find_elements(By.XPATH,'//span[@class="a-size-small a-color-link a-text-normal"]')
        return_exchange.append(re.text)# Getting the product return/exchange
    except NoSuchElementException :
        return_exchange.append('-')
        
    try:
        ed=amazon.find_elements(By.XPATH,'//span[@class="a-text-bold"]')
        expected_delivery.append(ed.text)# Getting the product expected delivery
    except NoSuchElementException :
        expected_delivery.append('-')
        
    try:
        avail=amazon.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        availability.append(avail.text)# Getting the product availability
    except NoSuchElementException :
        availability.append('-')
        
# Creating a data frame       
df=pd.DataFrame({'Brand':brand,'Product':product_name,'Price':price,'Return/Exchange':return_exchange,'Expected Delivery':expected_delivery,'Availability':availability,'Product URL':urls})
df.to_csv('Cycle.csv')
Cycle=pd.read_csv('Cycle.csv')# Saving the DataFrame to a CSV file
Cycle
```

##### 9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location.
```python
hostelworld=webdriver.Chrome()
hostelworld.get("https://www.hostelworld.com/")# Initializing the WebDriver 
```

```python
loc=hostelworld.find_element(By.CLASS_NAME,"native-input")
loc.send_keys('London')
select=hostelworld.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[2]/div/ul/li[2]')
select.click()# Finding the search bar and enter city name
```

```python
search=hostelworld.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[5]/button[1]')
search.click()# Selecting the reference city
```

```python
Hostel_Name = []
distance = []
overall_review = []
rating=[]

 # Extracting relevant information
names = hostelworld.find_elements(By.XPATH,'//div[@class="property-info-header"]/div/span')
for i in names:
    Hostel_Name.append(i.text)

dist = hostelworld.find_elements(By.XPATH,'//span[@class="distance-description"]')
for i in dist:
    distance.append(i.text)
    
review= hostelworld.find_elements(By.XPATH,'//div[@class="property-info-header"]/div/div/div/span[2]')
for i in review:
    overall_review.append(i.text)

ra= hostelworld.find_elements(By.XPATH,'//div[@class="property-info-header"]/div/div/div[1]/span[1]')
for i in ra:
    rating.append(i.text)
```

```python
ho_url=[]
property_description=[]
facilities=[]
total_reviews=[]

url=hostelworld.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]')
for i in url:
    ho_url.append(i.get_attribute('href'))# Scraping image URLs
    
# Extracting the relevant information for each hostel
for i in ho_url:
    hostelworld.get(i)
    
    try:
        pd=hostelworld.find_element(By.XPATH,'//div[@class="content collapse-content"]')
        property_description.append(pd.text)
    except NoSuchElementException as e:
        property_description.append('-')
        
    try:
        fa=hostelworld.find_element(By.XPATH,'//ul[@class="facilities"]')
        facilities.append(fa.text)
    except NoSuchElementException as e:
        facilities.append('-')
         
    try:
        tr=hostelworld.find_element(By.XPATH,'//div[@class="reviews"]')
        total_reviews.append(tr.text)
    except NoSuchElementException as e:
        total_reviews.append('-')
```

```python
Hostel=pd.DataFrame({"Hostel Name":Hostel_Name,"Distance from City Centre":distance,"Ratings":rating,"Total Reviews":total_reviews,"Overall Review":overall_review,"Facilities":facilities,"Property Description":property_description})
Hostel # Creating a DataFrame  
```
