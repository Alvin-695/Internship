##### 1&2  Write a python program which searches all the product under a particular product from www.amazon.in. The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars. 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "BrandName", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
amazon=webdriver.Chrome()
amazon.get('http://www.amazon.in')
```

```python
product=amazon.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
product.send_keys(input('Enter product name:'))
search_button=amazon.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
search_button.click()
```

```python
urls=[]
for page in range(0,3):
    url=amazon.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        urls.append(i.get_attribute('href'))
next_button=amazon.find_elements(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[29]/div/div/span/a[3]')
next_button.click()
len(urls)
```

```python
brand=[]
product_name=[]
expected_delivery=[]
availability=[]
price=[]
return_exchange=[]


for i in urls:
    amazon.get(i)
    time.sleep(1)
    try:
        br=amazon.find_elements(By.XPATH,'//td[@class="a-span9"]')
        brand.append(br.text)
    except NoSuchElementException :
        brand.append('-')

    try:
        product=amazon.find_elements(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        product_name.append(product.text)
    except NoSuchElementException :
        product_name.append('-')

    try:
        pr=amazon.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
        price.append(pr.text)
    except NoSuchElementException :
        price.append('-')
        
    try:
        re=amazon.find_elements(By.XPATH,'//span[@class="a-size-small a-color-link a-text-normal"]')
        return_exchange.append(re.text)
    except NoSuchElementException :
        return_exchange.append('-')
        
    try:
        ed=amazon.find_elements(By.XPATH,'//span[@class="a-text-bold"]')
        expected_delivery.append(ed.text)
    except NoSuchElementException :
        expected_delivery.append('-')
        
    try:
        avail=amazon.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        availability.append(avail.text)
    except NoSuchElementException :
        availability.append('-')
        
df=pd.DataFrame({'Brand':brand,'Product':product_name,'Price':price,'Return/Exchange':return_exchange,'Expected Delivery':expected_delivery,'Availability':availability,'Product URL':urls})
df.to_csv('Cycle.csv')
Cycle=pd.read_csv('Cycle.csv')
Cycle
```

##### 4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
flipkart=webdriver.Chrome()
flipkart.get('http://flipkart.com')
```

```python
search=flipkart.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input')
search.send_keys(input('Mobile brand:'))
search_button= flipkart.find_element(By.CLASS_NAME,'_2iLD__')
search_button.click()
```

```python
mobile_urls =[]
urls = flipkart.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for i in urls:
    mobile_urls.append(i.get_attribute("href"))
```

```python
Brand = []
Phone_name = []
Colour = []
RAM= []
ROM = []
Primary_Camera= []
Secondary_Camera = []
Display_Size= []
Battery_Capacity = []
Price = []

for url in mobile_urls:
    flipkart.get(url)              
   
    try:
        brand_tags = flipkart.find_element(By.XPATH,'//span[@class="B_NuCI"]')      
        Brand.append(brand_tags.text.split()[0])
    except NoSuchElementException:
        Brand.append('-')
    
    try:
        name_tags = flipkart.find_element(By.XPATH,'//div[@class="_3k-BhJ"][1]/table/tbody/tr[3]/td[2]/ul/li')     
        Phone_name.append(name_tags.text)
    except NoSuchElementException:
        Phone_name.append('-')
         
    try:
        ram_tags = flipkart.find_element(By.XPATH,'//div[@class="_3k-BhJ"][4]/table[1]/tbody/tr[2]/td[2]/ul/li')                
        RAM.append(ram_tags.text)
    except NoSuchElementException:
        RAM.append('-')
    
    try:
        rom_tags = flipkart.find_element(By.XPATH,'//div[@class="_3k-BhJ"][4]/table[1]/tbody/tr[1]/td[2]/ul/li')        
        ROM.append(rom_tags.text)
    except NoSuchElementException:
        ROM.append('-')
        
    try:                                                                                    
        pri_tags = flipkart.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div/div/div[2]/ul/li[3]')
        Primary_Camera.append(pri_tags.text)
    except NoSuchElementException:
        Primary_Camera.append('-')
          
    try:                                                                                    
        dis = flipkart.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div/div/div[2]/ul/li[2]')
        Display_Size.append(dis.text)
    except NoSuchElementException:
        Display_Size.append('-')
    
    try:                                                                                    
        col = flipkart.find_element(By.XPATH,'//div[@class="_3k-BhJ"][1]/table/tbody/tr[4]/td[2]/ul/li')
        Colour.append(col.text)
    except NoSuchElementException:
        Colour.append('-')
        
    try:                                                                                    
        batt = flipkart.find_element(By.XPATH,'//div[@class="_3k-BhJ"][7]/table/tbody/tr/td[2]/ul/li')
        Battery_Capacity.append(batt.text)
    except NoSuchElementException:
        Battery_Capacity.append('-')
    
    try:                                                                                    
        prices = flipkart.find_element(By.XPATH,'//div[@class="_30jeq3 _16Jk6d"]')
        Price.append(prices.text)
    except NoSuchElementException:
        Price.append('-')
        
df=pd.DataFrame({"brand_name":Brand,"smartphone_name":Phone_name,"product_colour":Colour,"product_RAM":RAM,"product_ROM":ROM,"Primary_Camera":Primary_Camera,"Display_size":Display_Size,"Battery_Capacity":Battery_Capacity,"Phone_Price":Price,"Product_URL":mobile_urls})
df.to_csv('onepluscsv')
df
```
