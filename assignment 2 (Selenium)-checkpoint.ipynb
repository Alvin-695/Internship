##### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
# opening chrome
driver=webdriver.Chrome()
```

```python
# opening the shine page on chrome through selenium
driver.get("https://www.shine.com/")
```

```python
# entering designation in search bar
designation=driver.find_element(By.CLASS_NAME,"form-control")
designation.send_keys("Data Analyst")
```

```python
# entering location in search bar
location=driver.find_element(By.ID,"id_loc")
location.send_keys('Bangalore')
```

```python
# searching the above input though automated instruction 
search=driver.find_element(By.CLASS_NAME,"searchForm_btnWrap_advance__VYBHN")
search.click()
```

```python
# making empty list for our data to store
job_title=[]
company_name=[]
experience=[]
location=[]

# scraping job title from the given page
title_tags=driver.find_elements(By.XPATH,'//h2[/html/body/div[1]/div[1]/div[4]/div/div[2]/div[1]/div/div/div[1]/div[1]/div[1]/h2]/a')
for i in title_tags:
    ti=i.text
    job_title.append(ti)

# scraping company name from the given page
co_tags=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in co_tags:
    co=i.text
    company_name.append(co)

# scraping experience from the given page
ex_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in ex_tags:
    ex=i.text
    experience.append(ex)

# scraping location from the given page
lo_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in lo_tags:
    lo=i.text
    location.append(lo)
    
# presenting the data in chart
df=pd.DataFrame({"Title":job_title,"Company":company_name,"Experience":experience,"Location":location})
df['Location'] = df['Location'].str.replace('\n\+\d+', '')
df[:11]
```

##### Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
   Brand
   Product Description
   Price

```python
flipkart=webdriver.Chrome() # opening chrome
```

```python
flipkart.get('https://www.flipkart.com/') # opening the flipkart page on chrome through selenium
```

```python
sunglasses=flipkart.find_element(By.CLASS_NAME,"Pke_EE")
sunglasses.send_keys("Sunglasses") # entering sunglasses in search bar
```

```python
search=flipkart.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button')
search.click() # searching the above input though automated instruction
```

```python
# making empty list for our data to store & giving range to page search
brand=[]
description=[]
price=[]
start=0
end=3

# scraping brand name from the given page and next 3 pages
for page in range(start,end):
    brand_name=flipkart.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in brand_name:
        brand.append(i.text)
    next_button=flipkart.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a/span')
    next_button.click()
    time.sleep(3)
    
# scraping description from the given page and next 3 pages    
for page in range(start,end):
    des=flipkart.find_elements(By.XPATH,'//div[@class="_2B099V"]/a[1]')
    for i in des:
        description.append(i.text)
    next_button=flipkart.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a/span')
    next_button.click()
    time.sleep(3)
# scraping price from the given page and next 3 pages
for page in range(start,end):
    rs=flipkart.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in rs:
        price.append(i.text)
    next_button=flipkart.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a/span')
    next_button.click()
    time.sleep(3)

# presenting the data in chart
df=pd.DataFrame({"Brand":brand,"Description":description,"Price":price})
df[0:100]
```

##### Q5: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link: https://www.flipkart.com/apple-iphone-11-black-64-gb/product- reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market place=FLIPKART

```python
flipkart=webdriver.Chrome() # opening chrome
```

```python
 # opening the flipkart apple mobile review page on chrome through selenium
flipkart.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART.") 
```

```python
# making empty list for our data to store & giving range to page search
rating=[]
summary=[]
full_review=[]
start=0
end=10

# scraping rating from the given page & next pages
for page in range(start,end):
    rt=flipkart.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for i in rt:
        rating.append(i.text)
    next_button=flipkart.find_element(By.XPATH,'//nav[@class="yFHi8N"]/a/span')
    next_button.click()
    time.sleep(3)
    
# scraping summary from the given page & next pages
for page in range(start,end):
    summ=flipkart.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for i in summ:
        summary.append(i.text)
    next_button=flipkart.find_element(By.XPATH,'//nav[@class="yFHi8N"]/a/span')
    next_button.click()
    time.sleep(3)

# scraping review from the given page & next pages
for page in range(start,end):
    re=flipkart.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for i in re:
        full_review.append(i.text.strip())
    next_button=flipkart.find_element(By.XPATH,'//nav[@class="yFHi8N"]/a/span')
    next_button.click()
    time.sleep(3)

# presenting the data in chart
df=pd.DataFrame({"Rating":rating,"Review Summary":summary,"Full Review":full_review})
df[0:100]
```

##### Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on Top Quotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
quotes=webdriver.Chrome()  # opening chrome
```

```python
quotes.get("https://www.azquotes.com/") # opening the azquotes page on chrome through selenium
```

```python
top=quotes.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
top.click() # selecting top quote from the page
```

```python
# making empty list for our data to store
quote=[]
author=[]
types=[]

# scraping quote from the given page
qot=quotes.find_elements(By.XPATH,'//a[@class="title"]')
for i in qot:
    quote.append(i.text)

# scraping author from the given page
auth=quotes.find_elements(By.XPATH,'//div[@class="author"]')
for i in auth:
    author.append(i.text)
    
# scraping type of quote from the given page
ty=quotes.find_elements(By.XPATH,'//div[@class="tags"]')
for i in ty:
    types.append(i.text)
    
# presenting the data in chart
df=pd.DataFrame({"Quote":quote,"Author":author,"Type of quote":types})
df[0:100]
```

##### Q9: Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/.

```python
josh=webdriver.Chrome() # opening chrome
```

```python
josh.get("https://www.jagranjosh.com/")  # opening the jagranjosh page on chrome through selenium
```

```python
gk=josh.find_element(By.XPATH,'/html/body/div[1]/div[19]/section[1]/div/ul/li[1]/span/a')
gk.click() # selecting gk section in the page
```

```python
pm=josh.find_element(By.XPATH,'/html/body/div[1]/div[8]/section[5]/div[2]/ul/li[4]/article/h3/a')
pm.click() # selecting prime minister page 
```

```python
# making empty list for our data to store
name=[]
born_dead=[]
term=[]
remark=[]

# scraping pm name from the given page
pm=josh.find_elements(By.XPATH,'//td[2]//div[/html/body/div[1]/main/div[1]/div[1]/article/div[3]/div[5]/div/table/tbody/tr[5]/td[2]/div]')
for i in pm:
    name.append(i.text)

# scraping born and dead from the given page
bd=josh.find_elements(By.XPATH,'//td[3]//div[/html/body/div[1]/main/div[1]/div[1]/article/div[3]/div[5]/div/table/tbody/tr[2]/td[3]/div]')
for i in bd:
    born_dead.append(i.text)
    
# scraping term from the given page
tm=josh.find_elements(By.XPATH,'//td[4][/html/body/div[1]/main/div[1]/div[1]/article/div[3]/div[5]/div/table/tbody/tr[2]/td[4]]')
for i in tm:
    term.append(i.text.strip())
    
# scraping remark from the given page
re=josh.find_elements(By.XPATH,'//td[5][/html/body/div[1]/main/div[1]/div[1]/article/div[3]/div[5]/div/table/tbody/tr[2]/td[5]/div]')
for i in re:
    remark.append(i.text)
    
# presenting the data in chart
df=pd.DataFrame({"PM Name":name,"Born-Dead":born_dead,"Term":term,"Remarks":remark})
df['Term'] = df['Term'].str.replace('\n', ', ')
df
```
